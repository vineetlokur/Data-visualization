```python
import numpy as np
x = np.arange(0,5,0.1)
y = x **3
#plt.plot(x,y,color ="black",marker="p")
plt.plot(x,y,color ="black",linewidth =5)

plt.plot()
plt.ylabel("X")
plt.xlabel("y")
plt.title("matplotlib demo")
plt.show()
```

```python
import numpy as np
x = np.arange(0,5,0.1)
y = x **3
y2 = x**2
#plt.plot(x,y,color ="black",marker="p")
plt.plot(x,y,color ="black",linewidth =5)
plt.plot(x,y2,color ="red",linewidth =5)

plt.plot()
plt.ylabel("X")
plt.xlabel("y")
plt.title("matplotlib demo")
plt.show()
```

```python
import numpy as np
x = np.arange(0,5,0.1)
y = x **3
y2 = x**2
#plt.plot(x,y,color ="black",marker="p")
plt.plot(x,y,color ="black",linewidth =5,label = "x^3")
plt.plot(x,y2,color ="red",linewidth =5,label ="x^2")

plt.plot()
plt.ylabel("X")
plt.xlabel("y")
plt.grid()

plt.text(2,80,"Graph",fontsize=12)
plt.legend()
plt.title("matplotlib demo")
plt.show()
```

```python
#bubble graph

year =[2012,2013,2014,2015,2016,2017]
salary =[12,13,14,15,16,20]
population =[100,120,180,250,300,370]
```

```python
plt.scatter(year,salary,s=100)
#plt.title("bubble chart")
plt.show()
```

```python
plt.scatter(year,salary,s=population,c="r")
#plt.title("bubble chart")
plt.show()
```

```python
plt.scatter(year,salary,s=population,c="r",alpha=0.1)
#plt.title("bubble chart")
plt.show()
```

```python
import numpy as np
c = np.arange(len(year))
```

```python
plt.scatter(year,salary,s=population,c=c)
#plt.title("bubble chart")
plt.show()
```

```python
plt.scatter(year,salary,s=population,c=c,alpha=0.5,edgecolor="black")
#plt.title("bubble chart")
plt.show()
```

```python
plt.scatter(year,salary,s=population,c=c,alpha=0.5,edgecolor="black",marker ='^')
plt.title("Bubble sort")
plt.text(year[0]+0.1,salary[0] + 0.1,population[0])
plt.text(year[2]+0.1,salary[2] + 0.1,population[2])
#plt.title("bubble chart")
plt.show()
```

```python
#pie chart

sizes =[23,3,4,6]
plt.pie(sizes)
plt.axis("equal")
plt.show()
```

```python
sizes =[23,3,4,6]
colors = ["blue","purple","red","pink"]
plt.pie(sizes,colors=colors)
plt.axis("equal")
plt.show()
```

```python
labels =["A","B","c","D"]
sizes =[23,3,4,6]
colors = ["blue","purple","red","pink"]
plt.pie(sizes,colors=colors,labels = labels,autopct ="%.2f")
plt.axis("equal")

plt.show()
```

```python
#add percentage
labels =["A","B","c","D"]
sizes =[23,3,4,6]
colors = ["blue","purple","red","pink"]
plt.pie(sizes,colors=colors,labels = labels,autopct ="%.2f%%")
plt.axis("equal")
plt.show()
```

```python
labels =["A","B","c","D"]
sizes =[23,3,4,6]
colors = ["blue","purple","red","pink"]
explode =[0.1,0.2,0,0]
plt.pie(sizes,colors=colors,explode = explode,labels = labels,autopct ="%.2f%%")
plt.axis("equal")
plt.show()
```
